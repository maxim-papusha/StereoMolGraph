```python
import py3Dmol
```

```python
# Twistane XYZ geometry
xyz_twistane = """26
Twistane C10H16
C   0.000000   1.214790   0.702710
C   0.000000  -1.214790   0.702710
C   1.214790   0.000000  -0.702710
C  -1.214790   0.000000  -0.702710
C   0.702710   1.214790  -0.000000
C  -0.702710  -1.214790  -0.000000
C   1.214790  -0.702710   0.000000
C  -1.214790   0.702710   0.000000
C   0.702710   0.000000   1.214790
C  -0.702710   0.000000  -1.214790
H   0.000000   2.150000   1.250000
H   0.000000  -2.150000   1.250000
H   2.150000   0.000000  -1.250000
H  -2.150000   0.000000  -1.250000
H   1.250000   2.150000   0.000000
H  -1.250000  -2.150000   0.000000
H   2.150000  -1.250000   0.000000
H  -2.150000   1.250000   0.000000
H   1.250000   0.000000   2.150000
H  -1.250000   0.000000  -2.150000
H   0.650000   1.650000  -0.950000
H  -0.650000  -1.650000  -0.950000
H   1.650000  -0.650000   0.950000
H  -1.650000   0.650000   0.950000
H   0.950000   0.650000   1.650000
H  -0.950000  -0.650000  -1.650000
"""


def mirror_xyz(xyz_str, axis="x"):
    """
    Return a mirrored XYZ string by reflecting all coordinates
    through the plane perpendicular to the chosen axis.
    axis='x' means x -> -x.
    """
    axis_to_idx = {"x": 0, "y": 1, "z": 2}
    idx = axis_to_idx[axis]

    lines = xyz_str.strip().splitlines()
    n_atoms = int(lines[0])
    comment = lines[1] + f" | mirrored along {axis}"

    mirrored_lines = [str(n_atoms), comment]

    for line in lines[2:]:
        parts = line.split()
        elem = parts[0]
        coords = [float(parts[1]), float(parts[2]), float(parts[3])]
        coords[idx] *= -1.0
        mirrored_lines.append(
            f"{elem:2s} {coords[0]: .6f} {coords[1]: .6f} {coords[2]: .6f}"
        )

    return "\n".join(mirrored_lines)


xyz_mirror = mirror_xyz(xyz_twistane, axis="x")

# Two-panel viewer
viewer = py3Dmol.view(width=900, height=420, viewergrid=(1, 2))

# Left: original twistane
viewer.addModel(xyz_twistane, "xyz", viewer=(0, 0))
viewer.setStyle({"stick": {}, "sphere": {"scale": 0.28}}, viewer=(0, 0))

# Right: mirrored coordinate set
viewer.addModel(xyz_mirror, "xyz", viewer=(0, 1))
viewer.setStyle({"stick": {}, "sphere": {"scale": 0.28}}, viewer=(0, 1))
viewer.zoomTo()
viewer.show()
```
